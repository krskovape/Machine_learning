### **Exercise 1:** line plot
* Plot line graph with plt.plot() method using x and y1 and y2 data
* Plot sin with blue line and cosfunction with green dashed line 
* Add plot label and legend 

```python
# Sample data
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# create plot 
plt.plot(x,y1,'b-', label="sinus")
plt.plot(x,y2,'g--', label="cosinus")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="best")
```

### **Exercise 2:**  scatterplot and line together

* Combine `plt.scatterplot()` and `plt.plot()`.  

* Use scatterplot for x and y1 samples; and lineplot for x and y2 for the 'underlying function'. 

* Add legend

```python
# Data
N = 50
x = np.linspace(0., 10., N)
y1 = np.sin(x)**2 + np.cos(x) + np.random.randn(N) / 10
y2 = np.sin(x)**2 + np.cos(x)
```

```python
# create plot 
plt.scatter(x,y1,c="b",label="x-y1 samples")
plt.plot(x,y2,"g-", label="line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc=0)
# plt.legend(["x-y1 samples", "line"])
```

### Exercise 3: multiplot

* Use subplot: nrows=1, ncols=2 to create the multiplot. 

* Plot (x,y) and (x,z) on the axes of 2 plot side by side. 



```python
# data 
x = np.arange(0,6)
y = x*2
z = x**2
```

```python
# create, multi-plot 
plt.subplot(1,2,1)
plt.plot(x,y,"bo-")
plt.xlabel("x")
plt.ylabel("y")
plt.subplot(1,2,2)
plt.plot(x,z,"g--")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout()
```
